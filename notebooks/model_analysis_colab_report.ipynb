````python
from pathlib import Path

# Change this path if your folder name/location is different.
PROJECT_DIR = Path('/content/drive/MyDrive/music-genre-classification')
REPORT_DIR = PROJECT_DIR / 'reports' / 'model_analysis'

if not REPORT_DIR.exists():
    raise FileNotFoundError(
        f'Report folder not found: {REPORT_DIR}\n'
        'Upload reports/model_analysis to Google Drive or update PROJECT_DIR.'
    )

print(f'Reading reports from: {REPORT_DIR}')
````

## 3. Metrics Summary

````python
import pandas as pd
from IPython.display import display

metrics_df = pd.read_csv(REPORT_DIR / 'model_metrics_summary.csv')
display(
    metrics_df.style
    .format({
        'accuracy': '{:.3f}',
        'precision_macro': '{:.3f}',
        'recall_macro': '{:.3f}',
        'f1_macro': '{:.3f}',
        'f1_weighted': '{:.3f}',
    })
    .background_gradient(subset=['accuracy', 'precision_macro', 'recall_macro', 'f1_macro', 'f1_weighted'], cmap='YlGnBu')
)
````

## 4. Model Comparison Plots

````python
from IPython.display import Image, display

display(Image(filename=str(REPORT_DIR / 'model_comparison_f1_macro.png')))
display(Image(filename=str(REPORT_DIR / 'model_metrics_heatmap.png')))
````

## 5. Per-Class F1 Scores

````python
per_class_df = pd.read_csv(REPORT_DIR / 'per_class_f1_scores.csv')
display(per_class_df.head(20))
display(Image(filename=str(REPORT_DIR / 'per_class_f1_heatmap.png')))
````

## 6. Classification Reports

````python
report_files = sorted(REPORT_DIR.glob('classification_report_*.csv'))

for report_file in report_files:
    model_name = report_file.stem.replace('classification_report_', '').replace('_', ' ').title()
    print(f'\n{model_name}')
    display(pd.read_csv(report_file, index_col=0).round(3))
````

## 7. Confusion Matrices

````python
confusion_files = sorted(REPORT_DIR.glob('confusion_matrix_*.png'))

for image_file in confusion_files:
    model_name = image_file.stem.replace('confusion_matrix_', '').replace('_', ' ').title()
    print(model_name)
    display(Image(filename=str(image_file)))
````

## 8. Feature Importances

Feature importance is available for tree-based models: XGBoost, Extra Trees, and CatBoost.

````python
importance_images = sorted(REPORT_DIR.glob('feature_importance_*.png'))

for image_file in importance_images:
    model_name = image_file.stem.replace('feature_importance_', '').replace('_', ' ').title()
    print(model_name)
    display(Image(filename=str(image_file)))
````

````python
importance_csv_files = sorted(REPORT_DIR.glob('feature_importance_*.csv'))

for csv_file in importance_csv_files:
    model_name = csv_file.stem.replace('feature_importance_', '').replace('_', ' ').title()
    print(f'\nTop 15 features - {model_name}')
    display(pd.read_csv(csv_file).head(15).round(4))
````
